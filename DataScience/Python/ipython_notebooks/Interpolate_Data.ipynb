```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oxy_cytc = pd.read_csv('oxidized_cyt_c.txt', header=None, names=['freq', 'abs'])
```

```python
red_cytc = pd.read_csv('reduced_cyt_c.txt', header=None, names=['freq', 'abs'])
```

```python
from scipy.interpolate import interp1d
```

```python
f_oxy = interp1d(oxy_cytc['freq'], oxy_cytc['abs'], kind='linear')
```

```python
f_red = interp1d(red_cytc['freq'], red_cytc['abs'], kind='linear')
```

```python
freq_range = np.linspace(250, 575)
```

```python
plt.plot(freq_range, f_oxy(freq_range), 'r-',
         freq_range, f_red(freq_range), 'b-'); plt.show()
```

```python
plt.rcParams['figure.figsize'] = 10, 6
plt.plot(oxy_cytc['abs'], 'r-',
         red_cytc['abs'], 'b-',
         f_oxy(freq_range), 'ro',
         f_red(freq_range), 'bo'
         );
plt.show()
```
